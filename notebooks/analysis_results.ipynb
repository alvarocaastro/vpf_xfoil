## Loading Results

```python
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Add src to path
project_root = Path().resolve().parent
sys.path.insert(0, str(project_root / "src"))

# Set style
plt.style.use('seaborn-v0_8')
%matplotlib inline
```

## Load Polar Data

```python
# Load polar data
results_dir = project_root / "results"
polars_dir = results_dir / "polars"

# Example: Load a specific polar
flight = "cruise"
section = "mid_span"
polar_file = polars_dir / f"{flight}_{section}.csv"

if polar_file.exists():
    df = pd.read_csv(polar_file)
    print(f"Loaded polar: {flight} / {section}")
    print(f"Data points: {len(df)}")
    print(df.head())
else:
    print(f"Polar file not found: {polar_file}")
```
